```python
import re
from collections import Counter

def top_k_words(filename: str, k: int) -> list:
    with open(filename, "r") as f:
        text = f.read().lower()


    words = re.findall(r'\w+', text)


    count = Counter(words)

    result =[]
    for word,freq in count.most_common(k):
        result.append(word)



    return result


with open("sample.txt", "w") as f:
    f.write("Data science is evolving. Python is versatile. Learning never stops.")

print(top_k_words("sample.txt", 3))
```

```python
def analyze_expenses():
    n = int(input("Enter number of transactions: "))

    expenses = {}

    for i in range(n):
        category, amount = input().split()
        amount = float(amount)


        if category in expenses:
            expenses[category] += amount
        else:
            expenses[category] = amount


    total = sum(expenses.values())


    highest_category = max(expenses, key=expenses.get)

    print("Total Expense:", total)
    print("Highest Category:", highest_category)
    print("Category-wise Total:", expenses)

analyze_expenses()
```

```python
def clean_data():
    n = int(input("Enter number of values: "))

    data = []

    for _ in range(n):
        value = input()

        try:
            num = float(value)


            if num >= 0:
                data.append(num)

        except ValueError:

            pass


        avg = sum(data) / len(data)
        max_val = max(data)


    print("Clean Data:", data)
    print("Average:", avg)
    print("Max:", max_val)

clean_data()
```
