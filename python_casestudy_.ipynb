```python
class bank_account:
   def __init__(self):
      self.balance=1000

   def deposit(self,amount):
    self.balance+=amount
    print("deposited:",amount)

   def withdraw(self,amount):
     if amount<=self.balance:
        self.balance-=amount
        print("withdraw:",amount)

     else:
          print("insufficient balance")

   def check_balance(self):
    print("current_balance =",self.balance)


obj = bank_account()


obj.deposit(500)
obj.withdraw(200)
obj.check_balance()
```

```python
class weatheranalyzer():

    def __init__(self,temp,humidity):
     self.temp=temp
     self.humidity=humidity

    def avg_temp(self):
      avg_temp=sum(self.temp)/len(self.temp)
      print("avg_temp=",avg_temp)

    def avg_humidity(self):
      avg_humidity=sum(self.temp)/len(self.humidity)
      print("avg_humidity=",avg_humidity)

    def weatheranalyzer(self):
      avg_temp=sum(self.temp)/len(self.temp)

      if avg_temp>30:
          print("weather condition :sunny")
      elif avg_temp >=25:
          print("weather condition :cloudy")
      else:
          print("weather condition :rainy")


temp=(30,25,28,39,20)
humidity=(70,80,56,72,50)

obj=weatheranalyzer(temp,humidity)
obj.avg_temp()
obj.avg_humidity()
obj.weatheranalyzer()








```
