```python
# Fill numerical columns with median
columns = [
    'Sale Price',
    'No of Bathrooms',
    'Flat Area (in Sqft)',
    'Lot Area (in Sqft)',
    'Area of the House from Basement (in Sqft)',
    'Latitude',
    'Longitude',
    'Living Area after Renovation (in Sqft)'
]

for col in columns:
    house[col] = house[col].fillna(house[col].median())
```

```python
house.isna().sum()
```

```python
house['Zipcode'] = house['Zipcode'].fillna(house['Zipcode'].mode()[0])
```

```python
freqgraph = house.select_dtypes(include=['int64','float64'])
freqgraph.hist(figsize=(20,20))
plt.show()
```

```python
num_cols = house.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    plt.figure(figsize=(6,4))
    plt.boxplot(house[col])
    plt.title(col)
    plt.show()
```

```python
num_cols = house.select_dtypes(include=['int64','float64']).columns

for col in num_cols:

    Q1 = house[col].quantile(0.25)
    Q3 = house[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    house = house[(house[col] >= lower) & (house[col] <= upper)]
```

```python
for col in num_cols:

    plt.figure(figsize=(6,4))
    plt.boxplot(house[col])
    plt.title(col)
    plt.show()
```

```python
house['Date House was Sold'] = pd.to_datetime(house['Date House was Sold'])

house['Year Sold'] = house['Date House was Sold'].dt.year
house['Month Sold'] = house['Date House was Sold'].dt.month
house['Day Sold'] = house['Date House was Sold'].dt.day

house.drop(columns=['Date House was Sold'], inplace=True)
```

```python
num_cols = house.select_dtypes(include=['int64','float64']).columns
```
